# **Key Rescaling Techniques**

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Sample data
data = np.array([[5], [10], [15], [20], [25]])

# Applying Min-Max Scaling
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

print("Original Data:\n", data)
print("Scaled Data:\n", scaled_data)
```

```python
from sklearn.preprocessing import StandardScaler

# Sample data
data = np.array([[5], [10], [15], [20], [25]])

# Applying Standardization
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data)

print("Original Data:\n", data)
print("Standardized Data (Z-score normalized):\n", standardized_data)
```

```python
from sklearn.preprocessing import RobustScaler

# Sample data with outliers
data = np.array([[5], [10], [15], [100], [25]])

# Applying Robust Scaling
scaler = RobustScaler()
robust_scaled_data = scaler.fit_transform(data)

print("Original Data:\n", data)
print("Robust Scaled Data:\n", robust_scaled_data)
```

```python
from sklearn.preprocessing import MaxAbsScaler

# Sample sparse-like data
data = np.array([[1], [2], [-4], [0], [5]])

# Applying MaxAbs Scaling
scaler = MaxAbsScaler()
maxabs_scaled_data = scaler.fit_transform(data)

print("Original Data:\n", data)
print("MaxAbs Scaled Data:\n", maxabs_scaled_data)
```

```python
from sklearn.preprocessing import PowerTransformer

# Sample data
data = np.array([[1], [2], [3], [4], [5]])

# Applying Yeo-Johnson Power Transformation (supports non-negative data)
scaler = PowerTransformer(method='yeo-johnson')
transformed_data = scaler.fit_transform(data)

print("Original Data:\n", data)
print("Power Transformed Data (Yeo-Johnson):\n", transformed_data)
```

```python
from sklearn.preprocessing import QuantileTransformer

# Sample data
data = np.array([[1], [2], [2], [3], [4]])

# Applying Quantile Transformation
scaler = QuantileTransformer(output_distribution='normal')
quantile_transformed_data = scaler.fit_transform(data)

print("Original Data:\n", data)
print("Quantile Transformed Data (Normal distribution):\n", quantile_transformed_data)
```

```python
import numpy as np

# Sample data
data = np.array([1, 10, 100, 1000, 10000])

# Applying Log Transform
log_transformed_data = np.log(data + 1)

print("Original Data:\n", data)
print("Log Transformed Data:\n", log_transformed_data)
```
